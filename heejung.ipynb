```python
import pandas as pd
import numpy as np
from rapidfuzz import process

# 1. (not set) -> (none) 치환
df['keyword'] = df['keyword'].replace('(not set)', '(none)')

# 2. 널값을 'not provided'로 채우기
df['keyword'] = df['keyword'].fillna('not provided')

# 3. +YouTube, +Google 등으로 시작하는 값들을 벡터화로 빠르게 치환
conditions = [
    df['keyword'].str.lower().str.startswith('+youtube'),
    df['keyword'].str.lower().str.startswith('+google'),
    df['keyword'].str.lower().str.startswith('+android'),
    df['keyword'].str.lower().str.startswith('+chrome'),
]
choices = ['YouTube MD', 'Google MD', 'Android MD', 'Chrome MD']

df['keyword'] = np.select(conditions, choices, default=df['keyword'])

# 4. rapidfuzz로 유사값 치환 (score_cutoff=80 이상만)
choices_fuzzy = ['YouTube MD', 'Google MD', 'Android MD', 'Chrome MD']

def fuzzy_map_rapid(x):
    if not isinstance(x, str):
        return x
    best_match = process.extractOne(x, choices_fuzzy, score_cutoff=80)
    return best_match[0] if best_match else x

df['keyword_cleaned'] = df['keyword'].apply(fuzzy_map_rapid)

# 5. 결과 확인
unique_keywords = sorted(df['keyword_cleaned'].dropna().unique())
pd.set_option('display.max_rows', None)
for keyword in unique_keywords:
    print(keyword)
```
